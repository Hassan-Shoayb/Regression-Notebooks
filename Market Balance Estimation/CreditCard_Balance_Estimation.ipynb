# Training and Testing the Algorithm

```python
# Splitting the data into Training and Testing Set
target = credit_data.Balance
features = credit_data.drop(["Balance"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(features, target,  test_size=0.2, random_state=10)

# Calculation the Percentage of Training and Testing
train_percent = len(X_train)/len(features)
test_percent = len(X_test)/len(features)

print("Training Set:", train_percent)
print("Testing Set:", test_percent)
```

```python
# Running Linear Regression
regr = LinearRegression()
regr.fit(X_train, y_train)

print("Training data R-SQUARED", regr.score(X_train, y_train))
print("Testing data R-SQUARED", regr.score(X_test, y_test))
print("Intercept:", regr.intercept_)
```

```python
DataFrame(regr.coef_, columns=["coef"], index=features.columns)
```

# P-values and Evaluating Coefficients 

```python
# Original Model Including All features
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, x_incl_const)
results = model.fit()

orig_pvalues = DataFrame(data={"coef":results.params, "p-values": round(results.pvalues, 3)})
orig_pvalues
```

# Testing for Multicollinearity

### $$  Age = \alpha _0 + \alpha _1Income + \alpha _2Studnt + ....+ \alpha  _9Married $$
### $$  VIF_{Age} = \frac{1}{(1 - R^2_{Age})} $$

```python
variance_inflation_factor(exog=x_incl_const.values, exog_idx=2)
```

```python
# Calculating the VIF of all the features
vif = []
for i in range(x_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog=x_incl_const.values, exog_idx=i))
    
orig_vif = DataFrame(data={"vif":vif}, index=x_incl_const.columns)

orig_frame = [orig_pvalues, np.around(orig_vif, 3)]
original_model = pd.concat(orig_frame, axis=1)

original_model
```

# Model Simplification and the Baysian Information Criterion (BIC)

```python
# Original Model and all features
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, x_incl_const)
results = model.fit()

orig_model = DataFrame({"coef": results.params, "p-values": round(results.pvalues, 3)})

print("R-squared:", results.rsquared)
print("BIC:", results.bic)
```

```python
# Model Excluding Education, Gender, Married due to their P-Values us greater than the threshhold
x_incl_const = sm.add_constant(X_train)
x_incl_const = x_incl_const.drop(["Education", "Gender", "Married", "Rating"], axis=1)
model = sm.OLS(np.log(y_train), x_incl_const)
results = model.fit()

excl_model = DataFrame({"coef": results.params, "p-value": round(results.pvalues, 3)})

print("R-squared:", results.rsquared)
print("BIC:", results.bic)
```

```python
_vif = []

for i in range(x_incl_const.shape[1]):
    _vif.append(variance_inflation_factor(exog=x_incl_const.values, exog_idx=i))
    
new_vif = DataFrame(data={"vif":np.around(_vif, 3)}, index=x_incl_const.columns)

new_frame = [excl_model, new_vif]
excl_features = pd.concat(new_frame, axis=1)
excl_features
```

```python
frames = [original_model, excl_features]
pd.concat(frames, axis=1)
```

# Residuals and residual Plots

```python
# Actual Vs Predicted Values Original Model and all features
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, x_incl_const)
results = model.fit()

plt.figure(figsize=[12, 8])
plt.scatter(y_train, results.fittedvalues)
plt.title(f"Actual Vs Fitted Values $y_i$ - $\hat y$ corr({round(y_train.corr(results.fittedvalues), 3)})")
plt.xlabel("Actual values $ y _i$", fontsize=14, color="blue")
plt.ylabel("Fitted values $ \hat y $", fontsize=14, color="blue")

plt.show()

# Predicted Values Vs Residuals 
plt.figure(figsize=[12, 8])
plt.scatter(results.fittedvalues, results.resid)
plt.xlabel("Fittedvalues $\hat y$", fontsize=14, color="blue")
plt.ylabel("Residuals", fontsize=14, color="blue")

plt.show()

# Distribution of Residuals - Checking for Normallity
plt.figure(figsize=[12, 9])
sns.histplot(results.resid)
plt.title(f"Residual Distribution: skew({round(results.resid.skew(), 3)}), Mean({round(results.resid.mean(), 3)})")

plt.show()
```

```python
# Model Excluding Education, Gender, Married due to their P-Values us greater than the threshhold
x_incl_const = sm.add_constant(X_train)
x_incl_const = x_incl_const.drop(["Education", "Gender", "Married", "Rating"], axis=1)
model = sm.OLS(np.log(y_train), x_incl_const)
results = model.fit()

plt.figure(figsize=[12, 8])
plt.scatter(y_train, results.fittedvalues)
plt.title(f"Actual Vs Fitted Values $y_i$ - $\hat y$ corr({round(y_train.corr(results.fittedvalues), 3)})")
plt.xlabel("Actual values $ y _i$", fontsize=14, color="blue")
plt.ylabel("Fitted values $ \hat y $", fontsize=14, color="blue")

plt.show()

# Predicted Values Vs Residuals 
plt.figure(figsize=[12, 8])
plt.scatter(results.fittedvalues, results.resid)
plt.xlabel("Fittedvalues $\hat y$", fontsize=14, color="blue")
plt.ylabel("Residuals", fontsize=14, color="blue")

plt.show()

# Distribution of Residuals - Checking for Normallity
plt.figure(figsize=[12, 9])
sns.histplot(results.resid)
plt.title(f"Residual Distribution: skew({round(results.resid.skew(), 3)}), Mean({round(results.resid.mean(), 3)})")

plt.show()
```
